```python
import os
from dotenv import load_dotenv
load_dotenv()

# Check if HF_TOKEN is loaded correctly
hf_token = os.getenv("HF_TOKEN")
if hf_token is None:
    print("HF_TOKEN is not set in the environment variables.")
else:
    print("HF_TOKEN loaded successfully.")

# Importing HuggingFaceEmbeddings correctly from LangChain
from langchain.embeddings import HuggingFaceEmbeddings

# Initialize Hugging Face embeddings model
embeddings = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")

# Set the Hugging Face token from environment variables if it exists
if hf_token:
    os.environ["HF_TOKEN"] = hf_token

# Check if embeddings were loaded correctly
print(embeddings)

```

```python
from pinecone_text.sparse import BM25Encoder
bm25_encoder=BM25Encoder().default()
bm25_encoder
```

```python
sentences=[
    "In  2023, I visited paris",
    "In 2021, I visited New York"

]
##tfidf values on these sentences
bm25_encoder.fit(sentences)
##store the values to json file
bm25_encoder.dump("bm25_values.json")
```

```python
retriever=PineconeHybridSearchRetriever(embeddings=embeddings,sparse_encoder=bm25_encoder,index=index)
retriever
```

```python
retriever.add_texts(
    [
        "In 2023, I visited paris",
        "In 2022, I visited New york ",
        "In 2021, I visited New Orleans",

    ]
)
```
